```python
import pandas

df = pandas.read_csv("Data_Comedian.csv")

print(df) 
```

Mengubah tipe data Nationality dan Go menjadi numerik

```python
d = {'UK': 0, 'USA': 1, 'N': 2}
df['Nationality'] = df['Nationality'].map(d)
d = {'YES': 1, 'NO': 0}
df['Go'] = df['Go'].map(d)

print(df) 
```

```python
features = ['Age', 'Experience', 'Rank', 'Nationality']

X = df[features]
y = df['Go']

print(X)
print(y) 
```

```python
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt

dtree = DecisionTreeClassifier()
dtree = dtree.fit(X, y)

tree.plot_tree(dtree, feature_names=features) 
```

```python
import pandas
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt

df = pandas.read_csv("Data_Comedian.csv")

d = {'UK': 0, 'USA': 1, 'N': 2}
df['Nationality'] = df['Nationality'].map(d)
d = {'YES': 1, 'NO': 0}
df['Go'] = df['Go'].map(d)

features = ['Age', 'Experience', 'Rank', 'Nationality']

X = df[features]
y = df['Go']

dtree = DecisionTreeClassifier()
dtree = dtree.fit(X, y)

tree.plot_tree(dtree, feature_names=features)
```

```python
import pandas
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt

df = pandas.read_csv("Jaringan_Komunikasi.csv")
# print(df)

d = {'Pendek': 0, 'Panjang': 1}
df['Waktu'] = df['Waktu'].map(d)
d = {'Besar': 1, 'Kecil': 0}
df['Paket'] = df['Paket'].map(d)
d = {'Rendah': 0, 'Sedang': 1, 'Tinggi':2}
df['Frekuensi'] = df['Frekuensi'].map(d)
d = {'Rendah': 0, 'Tinggi': 1}
df['Prioritas'] = df['Prioritas'].map(d)
d = {'Gangguan': 0, 'Normal': 1}
df['Gangguan'] = df['Gangguan'].map(d)

# print(df)

features = ['Waktu', 'Paket', 'Frekuensi', 'Prioritas']

X = df[features]
y = df['Gangguan']

dtree = DecisionTreeClassifier()
dtree = dtree.fit(X, y)

tree.plot_tree(dtree, feature_names=features)
```
